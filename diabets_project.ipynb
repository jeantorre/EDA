```python
glucose_mean = round(df_diabetes['Glucose'].mean(), 2)
bloodpressure_mean = round(df_diabetes['BloodPressure'].mean(), 2)
bmi_mean = round(df_diabetes['BMI'].mean(), 2)
skin_median = round(df_diabetes['SkinThickness'].median(), 2)
insulin_median = round(df_diabetes['Insulin'].median(), 2)
```

```python
df_diabetes['Glucose'] = df_diabetes['Glucose'].replace(0, glucose_mean)
df_diabetes['BloodPressure'] = df_diabetes['BloodPressure'].replace(0, bloodpressure_mean)
df_diabetes['BMI'] = df_diabetes['BMI'].replace(0, bmi_mean)
df_diabetes['SkinThickness'] = df_diabetes['SkinThickness'].replace(0, skin_median)
df_diabetes['Insulin'] = df_diabetes['Insulin'].replace(0, insulin_median)
```

## ML Treatment

```python
df_diabetes.info()
```

```python
df_diabetes.columns
```

```python
x = df_diabetes[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
                 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = df_diabetes['Outcome']
```

```python
x.head(3)
```

```python
y.head(3)
```

```python
#Normalizing data

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x.values)
```

```python
#Splitting in train and test values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_reg = LogisticRegression()
logistic_reg.fit(x_train, y_train)
```

```python
y_pred = logistic_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

print('The accuracy from logistic regression is',accuracy_score(y_test, y_pred) * 100)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, fmt = '.2f', cbar = False);
plt.title('Confusion Matrix');
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
print(classification_report(y_test, y_pred))
```

### Random Forest

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(x_train, y_train)
```

```python
y_pred = forest.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, fmt = '.2f', cbar = False);
plt.title('Confusion Matrix');
```

```python
print(classification_report(y_test, y_pred))
```

### Neural network

```python
import tensorflow as tf
neural_network = tf.keras.models.Sequential()
neural_network.add(tf.keras.layers.Dense(units = 4, activation = 'relu', input_shape = (8,)))
neural_network.add(tf.keras.layers.Dense(units = 4, activation = 'relu'))
neural_network.add(tf.keras.layers.Dense(units = 4, activation = 'relu'))
neural_network.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
neural_network.summary()
```

```python
neural_network.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
neural_network.fit(x_train, y_train, epochs = 200)
```

```python
y_pred = neural_network(x_test)
y_pred = (y_pred >= .5)
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, fmt = '.2f', cbar = False);
plt.title('Confusion Matrix');
```

```python
print(classification_report(y_test, y_pred))
```
