```python
from sklearn.preprocessing import StandardScaler #padronização dos dados
```

```python
scaler = StandardScaler()
df_mkt_scaled = scaler.fit_transform(df_mkt)
```

```python
from sklearn.cluster import KMeans
k = range(1,21)

def calcular_wcss(dataframe):
    wcss = []
    for i in k:
        kmeans = KMeans(n_clusters = i, random_state = 0, init = 'k-means++')
        kmeans.fit(dataframe)
        wcss.append(kmeans.inertia_)
    return wcss
```

### Método cotovelo

```python
wcss = calcular_wcss(df_mkt_scaled)
```

```python
fig, ax = plt.subplots()
ax = sns.lineplot(x = k, y = wcss, marker = 'o')
plt.title('Número Clusters x WCSS', fontsize = 15)
ax.set_ylabel('WCSS', labelpad = 10, fontsize = 12)
ax.set_xlabel('Número Clusters', labelpad = 10, fontsize = 12)
plt.xticks(range(1, 21, 1))
plt.show()
```

```python
kmeans_mkt = KMeans(n_clusters = 8, random_state = 0, init = 'k-means++')
kmeans_mkt.fit(df_mkt_scaled)
rotulo = kmeans_mkt.labels_
np.unique(rotulo, return_counts = True)
```

```python
centro_cluster = pd.DataFrame(data = kmeans_mkt.cluster_centers_, columns = [df_mkt.columns])
centro_cluster
```

```python
centro_cluster = scaler.inverse_transform(centro_cluster)
centro_cluster = pd.DataFrame(data = centro_cluster, columns = [df_mkt.columns])
centro_cluster
```

#### Quantidade ótima de grupos (Fórmula)

```python
from math import sqrt

def numero_cluster_ideal(wcss):
    x1, y1 = k[0], wcss[0]
    x2, y2 = k[-1], wcss[-1]

    distancia = []
    for i in range(len(wcss)):
        x0 = i + 1
        y0 = wcss[i]
        numerador = abs((y2-y1)*x0 - (x2-x1)*y0 + x2*y1 - y2*x1)
        denominador = sqrt((y2 - y1)**2 + (x2 - x1)**2)
        distancia.append(numerador/denominador)
    
    return distancia.index(max(distancia)) + 1
```

```python
numero_cluster_ideal(wcss)
```
