## Regresso Logstica

```python
X_categorico = df_rh[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_categorico.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_categorico = onehotencoder.fit_transform(X_categorico).toarray()
```

```python
X_categorico = pd.DataFrame(X_categorico)
X_categorico.shape
```

```python
X_numerico = df_rh[['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction', 'HourlyRate',
                    'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
                    'OverTime', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel',
                    'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole',
                    'YearsSinceLastPromotion', 'YearsWithCurrManager']]
X_numerico.shape
```

```python
X_completo = pd.concat([X_categorico, X_numerico], axis = 1)
X_completo
```

 necessrio normalizar os dados para no acontecer de valores muito alto, como por exemplo a coluna 'DailyRate', ser considerado como mais importante do que valores mais baixos, como idade, educao entre outros.  
Dessa forma todos os valores vo assumir entre 0 e 1.

```python
type(X_completo.values)
```

```python
#Normalizando dados

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_completo.values)
```

```python
y = df_rh['Attrition']
```

```python
#Separao entre vlores treino e teste

from sklearn.model_selection import train_test_split
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = .2)
```

```python
print(X_treino.shape)
print(X_teste.shape)
```

```python
from sklearn.linear_model import LogisticRegression
reg_logistica = LogisticRegression()
reg_logistica.fit(X_treino, y_treino)
```

```python
y_pred = reg_logistica.predict(X_teste)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_teste, y_pred)
```

```python
#matriz confuso
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_teste, y_pred)
cm
```

```python
sns.heatmap(cm, annot = True, fmt = '.2f', cbar = False);
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
print(precision_score(y_teste, y_pred))
print(recall_score(y_teste, y_pred))
print(f1_score(y_teste, y_pred, average = 'macro'))
print(classification_report(y_teste, y_pred))
```

## Random Forest

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(X_treino, y_treino)
```

```python
y_pred = forest.predict(X_teste)
```

```python
accuracy_score(y_teste, y_pred)
```

```python
cm = confusion_matrix(y_teste, y_pred)
sns.heatmap(cm, annot = True, fmt = '.2f', cbar = False);
```

```python
print(classification_report(y_teste, y_pred))
```

## Redes Neurais

```python
import tensorflow as tf
rede_neural = tf.keras.models.Sequential()
rede_neural.add(tf.keras.layers.Dense(units = 25, activation = 'relu', input_shape = (50,)))
rede_neural.add(tf.keras.layers.Dense(units = 25, activation = 'relu'))
rede_neural.add(tf.keras.layers.Dense(units = 25, activation = 'relu'))
rede_neural.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
rede_neural.fit(X_treino, y_treino, epochs = 200)
```

```python
y_pred = rede_neural(X_teste)
y_pred[:10]
```

```python
y_pred = (y_pred >= .5)
y_pred[:10]
```

```python
cm = confusion_matrix(y_teste, y_pred)
sns.heatmap(cm, annot = True, fmt = '.2f', cbar = False);
```

```python
print(classification_report(y_teste, y_pred))
```
